### EM - Algorithm and Comparing with Kmeans

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sns
from sklearn import cluster, datasets, mixture
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
np.random.seed(12)
%matplotlib inline
```

```python
x_1 = sps.multivariate_normal(mean = [0,4],cov = [[1,0],[0,1]]).rvs(size = 1000)
x_2 = sps.multivariate_normal(mean = [-4,4] , cov = [[1,0],[0,1]]).rvs(size = 1000)
x_3 = sps.multivariate_normal(mean = [-2,-2] , cov = [[1,0],[0,1]]).rvs(size = 1000)
fig,ax = plt.subplots()
ax.scatter(x_1[:,0],x_1[:,1],color='magenta')
ax.scatter(x_2[:,0],x_2[:,1],color = 'darkgreen');
ax.scatter(x_3[:,0],x_3[:,1],color = 'red');
```

```python
features = np.concatenate([x_1,x_2,x_3])
gm = mixture.GaussianMixture(n_components=3,n_init = 30,init_params = 'random').fit(features)
y_pred = gm.predict(features)
colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00','#4daf4a']),int(max(y_pred) + 1))))
plt.scatter(features[:,0],features[:,1],color=colors[y_pred])
```

```python
n_samples = 2000
random_state = 42
noisy_circles = datasets.make_circles(n_samples = n_samples, factor = 0.5 ,noise = 0.05)
plt.scatter(noisy_circles[0][:,0],noisy_circles[0][:,1],color = 'magenta');
```

```python
features,target = noisy_circles[0],noisy_circles[1]
gm = mixture.GaussianMixture(n_components = 2,n_init =30,init_params = 'random').fit(features)
```

```python
y_pred = gm.predict(features)
```

```python
colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00']),int(max(y_pred) + 1))))
plt.scatter(noisy_circles[0][:,0],noisy_circles[0][:,1],color = colors[y_pred]);
```
